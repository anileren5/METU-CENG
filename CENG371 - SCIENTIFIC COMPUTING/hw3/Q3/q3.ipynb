#### part a)

```python
# Define N(t)
def N(t):
  return 0.3 + 2*t - 1.2*(t**2) + 0.5*(t**3)

# Define N_obs(t)
def N_obs(t):
  epsilons = np.random.normal(0, 0.01, 4)
  N_observed = 0.3 * 2*(t + epsilons[1]) - 1.2 * ((t + epsilons[2])**2) + 0.5 * ((t + epsilons[3])**3)
  return epsilons, N_observed
```

```python
# Define a function for generating set of observations
def O_n(n):
  observations = []
  for i in range(1, n+1):
    t = i / (n+1)
    epsilons, N_observed = N_obs(t)
    observations.append((t, N_observed, epsilons))
  return observations
```

```python
# Generate set of observations (O_n) and as well as A_n and b_n

# For n = 5
temp_5 = O_n(5)
O_5 = []
A_5 = []
b_5 = []
for element in temp_5:
  A_5.append(element[2])
  b_5.append(element[1])
  O_5.append((element[0], element[1]))


for i in range(len(A_5)):
  for j in range(4):
    if (j == 0):
      A_5[i][j] = 1
    elif (j == 1):
      A_5[i][j] = A_5[i][j]
    elif (j == 2):
      A_5[i][j] = (A_5[i][j]) ** 2
    elif (j == 3):
      A_5[i][j] = (A_5[i][j]) ** 3

print("------------------- O_5 -------------------")
for observation in O_5:
  print(observation)

print("\n ------------------- A_5 -------------------")
for row in A_5:
  print(row)

print("\n ------------------- b_5 -------------------")
for entry in b_5:
  print(entry)
```

#### part b)

I will use Augmented System techniques which is explained in lecture notes part 9.

```python
# For simplicity I will convert A_n's b_n's to numpy arrays

A_5 = np.array(A_5)
b_5 = np.array(b_5)

A_10 = np.array(A_10)
b_10 = np.array(b_10)

A_100 = np.array(A_100)
b_100 = np.array(b_100)
```

```python
# Define a function for augmented system construction
def BuildAugmentedSystem(A, b):
  # Make A and b numpy array if they are not
  A = np.array(A)
  b = np.array(b)

  # Dimensions
  m = A.shape[0]
  n = A.shape[1]

  # Construct I, A_T and O
  I = np.identity(m) # Identitiy matrix of size m*m
  A_T = A.transpose() # Transpose of A
  O = np.zeros((n,n)) # Zero matrix

  IA = np.hstack((I,A))
  A_TO = np.hstack((A_T, O))

  augmented_left = np.vstack((IA, A_TO))
  augmented_right = np.append(b, np.zeros((n)))

  return augmented_left, augmented_right
```

```python
# Estimation for c_k for n = 5 (A_n  * c_k = b_n)

# Construct augmented system for n = 5
augmented_left_5, augmented_right_5 = BuildAugmentedSystem(A_5, b_5)

# Solve augmented system for n = 5
r_x = solve(augmented_left_5, augmented_right_5)

# x = c_k_5
c_k_5 = r_x[-4:]

print(c_k_5)
```
